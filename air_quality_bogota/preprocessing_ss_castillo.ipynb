```python
import pandas as pd # libreria para el manejo y an√°lisis de estructuras de datos
import os, re # libreria que permite acceder a las funcionalidades dependientes del SO , libreria para el manejo de languajes regulares
```

```python
from os import walk # se utiliza para leer recursivamente los archivos y directorios del directorio indicado

f = []
for (dirpath, dirnames, filenames) in walk('data/raw/'):
    f.extend(filenames)
    break
f
```

```python
# se lee cado uno de los archivos y se agregan al DataFrame
path = os.path.abspath('data/raw/') 
files = os.listdir(path)  

df = pd.DataFrame()
for file in files:
    if file.endswith('.xlsx'):
        dft = pd.read_excel('data/raw/'+file)
        dft['Station'] = re.search(r'_(.*?)_', str(file)).group(1)
        df = df.append(dft[1:], ignore_index=True) 
df.head() 
```

```python
df.head() # mostras las 5 primeras filas, para comprobar los datos
```

```python
df.isnull().sum() # contamos los valores nulos
```

```python
df = df[['PM10','PM2.5','NO','NO2','NOX','CO','OZONO','Station', 'DateTime']] # se crea un nuevo DataFrame con las columnas especificadas
```

```python
df.head() # mostras las 5 primeras filas, para comprobar los datos
```

```python
df.columns # se muestran las columnas del DataFrame
```

```python
col = ['PM10', 'PM2.5', 'NO', 'NO2', 'NOX', 'CO', 'OZONO']
for i in col: # se recorre cada columna y se cambia su tipo de valor por uno numerico (float64) 
    df[i] = pd.to_numeric(df[i],errors='coerce')
```

```python
# contamos la cantidad de nulos en cada columna
df.isnull().sum()
```

```python
# cantidad de filas y columnas
df.shape
```

```python
# se reemplazan los valores nulos por la media
df['PM10'].fillna((df['PM10'].mean()), inplace=True)
#df['PM2.5'].fillna((df['PM2.5'].mean()), inplace=True)
df['CO'].fillna((df['CO'].mean()), inplace=True)
df['NO'].fillna((df['NO'].mean()), inplace=True)
df['NO2'].fillna((df['NO2'].mean()), inplace=True)
df['NOX'].fillna((df['NOX'].mean()), inplace=True)
df['OZONO'].fillna((df['OZONO'].mean()), inplace=True)
```

```python
# se verifica si hay valores nulos
df.isnull().sum()
```

```python
# Verdadero es mayor que 12, malo (Puro, No Puro)
df['Status'] = df['PM2.5']>12
df.Status.value_counts()
```

```python
# se guardan los datos en un nuevo archivo
df.to_csv("data/clean_data_final.csv", index = False)
```
